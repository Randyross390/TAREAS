```python
columna1 = sales['TOTAL_VALUE_SALES']
columna2 = sales['REGION']
columna3 = sales["ITEM_CODE"]
```

```python
# Selecciona solo columnas numéricas relevantes
X = sales[['TOTAL_UNIT_SALES', 'TOTAL_VALUE_SALES', 'TOTAL_UNIT_AVG_WEEKLY_SALES']]
```

```python
X = X.dropna()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# Aplicación de K-MEANS

```python
from sklearn.cluster import KMeans

wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init="k-means++", random_state=2)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

# Metodo codo de JAMBÚ

```python
plt.plot(range(1, 11), wcss)
plt.xlabel('Número de Clusters')
plt.ylabel('WCSS')
plt.title('Método del Codo')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init="k-means++", random_state=2)
clusters = kmeans.fit_predict(X_scaled)
X['Cluster'] = clusters
```

```python
# PREDICCION de cada grupo que toca o de donde debe de tocar este algoritmo
cluster = kmeans.fit_predict(X_scaled)
cluster
```

```python
plt.figure(figsize= (10,10)) # Creamos una imagen
plt.scatter(X_scaled[cluster==0,0], X_scaled[cluster==0,1], c="blue", label = "Cluster 1") # Graficamos todos los cluster del valor 0 y los puntos del primer cluster de color azul
plt.scatter(X_scaled[cluster==1,0], X_scaled[cluster==1,1], c="orange", label = "Cluster 2") # Graficamos todos los cluster del valor 1 y los puntos del primer cluster de color azul
plt.scatter(X_scaled[cluster==2,0], X_scaled[cluster==2,1], c="green", label = "Cluster 3") # Graficamos todos los cluster del valor 2 y los puntos del primer cluster de color verde
plt.scatter(X_scaled[cluster==3,0], X_scaled[cluster==3,1], c="yellow", label = "Cluster 4") # Graficamos todos los cluster del valor 3 y los puntos del primer cluster de color amarillo
plt.scatter(X_scaled[cluster==4,0], X_scaled[cluster==4,1], c="red", label = "Cluster 5") # Graficamos todos los cluster del valor 4 y los puntos del primer cluster de color red

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1], c = "black", label="Centroides") # Mostramos de negro todos los Centroides de los Clusters
plt.title("Grupos de clientes")
plt.xlabel = "Annual income"
plt.ylabel = "Spending Score"
plt.legend()
```

```python
pivot = sales.groupby(['Cluster', 'REGION'])['TOTAL_VALUE_SALES'].mean().unstack()
print(pivot)
```

```python
# Agrupa por semana y cluster, sumando las ventas
ventas_semanales = sales.groupby(['WEEK', 'Cluster'])['TOTAL_VALUE_SALES'].sum().reset_index()

# Pivot para que cada cluster sea una línea diferente
pivot = ventas_semanales.pivot(index='WEEK', columns='Cluster', values='TOTAL_VALUE_SALES')

# Gráfico
plt.figure(figsize=(16,8))
pivot.plot(ax=plt.gca(), marker='o')
plt.title('Ventas semanales por Cluster')
plt.xlabel('Semana')
plt.ylabel('Total de Ventas')
plt.legend(title='Cluster')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Ya tienes el DataFrame pivot de la opción anterior
pivot.plot(kind='bar', stacked=True, figsize=(18,9))
plt.title('Ventas semanales por Cluster (barras apiladas)')
plt.xlabel('Semana')
plt.ylabel('Total de Ventas')
plt.legend(title='Cluster')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

# Cluster 0 (Azul)
### Descripción: Es el cluster con mayores ventas semanales y muestra mucha estabilidad a lo largo del tiempo.
### Interpretación: Probablemente agrupa los productos/clientes más fuertes y constantes en ventas, los de mayor volumen o los clientes "premium"/más activos. Es el grupo más importante para el negocio en términos de facturación.
# Cluster 1 (Naranja)
### Descripción: Segundo cluster en importancia de ventas, también presenta un comportamiento bastante constante, aunque con algo más de variabilidad que el azul.
### Interpretación: Representa productos/clientes con buen desempeño pero un escalón por debajo del cluster azul. Pueden ser clientes regulares pero no top, o productos de buena rotación.
# Cluster 2 (Morado)
### Descripción: Presenta ventas intermedias, por debajo del naranja, pero con mayor volatilidad (subidas y bajadas notorias).
### Interpretación: Puede incluir clientes/productos con ventas estacionales o más dependientes de promociones/eventos, o bien con comportamiento menos predecible.
# Cluster 3 (Verde).
### Descripción: Ventas bajas y relativamente estables a lo largo del tiempo, con pequeñas variaciones.
### Interpretación: Engloba productos o clientes de baja rotación o menos relevantes, pero que mantienen cierto nivel de ventas de manera continua.
# Cluster 4 (Rojo)
### Descripción: Este cluster está prácticamente en cero ventas durante todo el periodo.
### Interpretación: Seguramente se trata de productos/clientes inactivos, descontinuados, o con ventas anecdóticas. Pueden representar outliers, errores o casos muy particulares.
### Resumen visual
# Cluster 0 y 1: Son los protagonistas del negocio, con ventas altas y consistentes.
# Cluster 2: Segmento intermedio, ventas medianas y comportamiento más variable.
# Cluster 3: Segmento bajo, ventas reducidas pero constantes.
# Cluster 4: Prácticamente sin ventas, inactivos o casos especiales.
