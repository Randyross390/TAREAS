```python
import pandas as pd
import numpy as np
```

```python
# Agregacion
dim_category = pd.read_csv('DIM_CATEGORY (2).csv')
dim_product = pd.read_excel('DIM_PRODUCT (1).xlsx')
dim_segment = pd.read_excel('DIM_SEGMENT (1).xlsx')
dim_calendar = pd.read_excel('DIM_CALENDAR (2).xlsx')
fact_sales = pd.read_csv('FACT_SALES (1).csv')
```

```python
# Arreglo
arregloDim_Categ = np.array(dim_category)
arregloDim_Produc = np.array(dim_product)
arregloDim_Seg = np.array(dim_segment)
arregloDim_Calen = np.array(dim_calendar)
arregloFact_sales = np.array(fact_sales)
```

```python
# DataFrame
df_dim_category = pd.DataFrame(arregloDim_Categ)
df_dim_product = pd.DataFrame(arregloDim_Produc)
df_dim_segment = pd.DataFrame(arregloDim_Seg)
df_dim_calendar = pd.DataFrame(arregloDim_Calen)
df_fact_sales = pd.DataFrame(arregloFact_sales)
```

# Realizar limpieza de datos

```python
# Eliminacion de duplicados
df_dim_category = df_dim_category.drop_duplicates()
df_dim_product = df_dim_product.drop_duplicates()
df_dim_segment = df_dim_segment.drop_duplicates()
df_dim_calendar = df_dim_calendar.drop_duplicates()
df_fact_sales = df_fact_sales.drop_duplicates()
```

```python
# Remplazo y Eliminacion de valores faltantes
df_dim_category = df_dim_category.fillna('Desconocido')
df_dim_product = df_dim_product.fillna('Desconocido')
df_dim_segment = df_dim_segment.fillna('Desconocido')
df_dim_calendar = df_dim_calendar.fillna('Desconocido')
df_fact_sales = df_fact_sales.fillna(0)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# Visualizar la distribución de ventas:

### Crea gráficos de distribución (histogramas o boxplots) para visualizar la distribución de ventas por diferentes categorías. Esto ayudará a identificar patrones de ventas y posibles outliers.

```python
manu_unit_sales = df_Prod_Sales.groupby('Manufacturer')['total_unit_sales'].mean()
```

```python
manu_unit_sales.plot(kind='bar')
plt.xlabel('Fabricante')
plt.ylabel('Total Unit Sales')
plt.title('Ventas Totales de Unidades por Fabricante')
```

```python
manu_value_sales = df_Prod_Sales.groupby('Manufacturer')['total_value_sales'].mean()
```

```python
manu_value_sales.plot(kind='bar')
plt.xlabel('Fabricante')
plt.ylabel('Total Value Sales')
plt.title('Total del valor de ventas por Fabricante')
```

```python
manu_avg_weekly_sales = df_Prod_Sales.groupby('Manufacturer')['total_unit_avg_weekly_sale'].mean()
```

```python
manu_avg_weekly_sales.plot(kind='bar')
plt.xlabel('Fabricante')
plt.ylabel('Total_unit_avg_weekly_Sales')
plt.title('Promedio total de unidades vendidas por semana')
```

# Analizar la tendencia de ventas en el tiempo:

### Genera gráficos de líneas que muestren las tendencias de ventas a lo largo del tiempo. Filtra las ventas por diferentes productos, regiones o segmentos para analizar cómo han cambiado las ventas a lo largo del tiempo.

```python
print(df_Prod_Sales.columns)
df_Prod_Sales['total_unit_sales'] = df_Prod_Sales['total_unit_sales'].round().astype(int)
df_Prod_Sales['total_value_sales'] = df_Prod_Sales['total_value_sales'].round().astype(int)
```

```python
plt.figure(figsize=(14, 7))
sns.lineplot(data=df_Prod_Sales, x='WEEK', y='total_unit_avg_weekly_sale', hue = 'Manufacturer', marker='o')
plt.xlabel('Semana')
plt.ylabel('Total Unit Sales')
plt.title('Tendencia de Ventas por Unidades a lo Largo del Tiempo')
plt.legend(title='Manufacturer')
plt.show()
```

```python
df_example = pd.DataFrame(df_Prod_Sales)
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=df_Prod_Sales,
    x="WEEK",
    y="total_unit_sales",
    hue="Manufacturer",
    marker="o"
)
plt.xlabel("Semana")
plt.ylabel("Ventas Totales por Unidades")
plt.title("Tendencia de Ventas a lo Largo del Tiempo por Fabricante")
plt.legend(title="Fabricante")
plt.grid(True)
plt.show()
```

# Explorar la relación entre diferentes variables:

### Utiliza gráficos de dispersión (scatter plots) para investigar relaciones entre variables clave. Esto puede ayudar a identificar correlaciones significativas entre diferentes variables.

```python
fig, ax = plt.subplots()
ax.scatter(df_Prod_Sales['total_value_sales'], df_Prod_Sales['total_unit_sales'], color = 'red', label = 'tota_unit_sales')
ax.scatter(df_Prod_Sales['total_value_sales'], df_Prod_Sales['Manufacturer'], color = 'blue', label = 'total_value_sales')
ax.set_xlabel(['total_unit_value_sales'])   
ax.set_ylabel('Manufacturer') 
ax.set_xticklabels(df_Prod_Sales['total_unit_sales'].index, rotation=90) 
ax.legend()
plt.show() 
```
